```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("TrainSchedulePipeline")\
.master("local[*]")\
.getOrCreate()
```

```python
import pandas as pd

df_pandas = pd.read_csv("/home/jovyan/work/Dataset1 (1).csv")

print(df_pandas.head())
```

```python
df = spark.read.csv("/home/jovyan/work/Dataset1 (1).csv",
                    header = True,
                    inferSchema = True
                   )
print("Dataset loaded ")
df.show(5)
```

```python
# Task 2.1 - Identify and Handel Missing Values

print("Missing Values in each column : ")
print(df_pandas.isnull().sum())

print("\nTotal Missing Values :")
print(df_pandas.isnull().sum().sum())
```

```python
# Task 2.2 - Detect and Remove Duplicate records

print("Total duplicate Rows : ")
print(df_pandas.duplicated().sum())

df_cleaned = df_pandas.drop_duplicates()   #remove duplicates

print("\nTotal Records before cleaning : ",len(df_pandas))
print("\nTotal records after cleaning : ",len(df_cleaned))
print("\nDuplicate removed Successfully ")
```

```python
# Task 2.3 - Standardize arrival and departure time formats

import pandas as pd

# - Convert to proper time format
df_cleaned['Arrival_time'] =  pd.to_datetime(
    df_cleaned['Arrival_time'],
    format = '%H:%M:%S'
).dt.strftime('%H:%M:%S')

df_cleaned['Departure_Time'] = pd.to_datetime(
    df_cleaned['Departure_Time'],
    format = '%H:%M:%S'
).dt.strftime('%H:%M:%S')

print("Time formats Standardized")
print("\nSample Arrival Time :")
print(df_cleaned['Arrival_time'].head())
print("\nSample Departure Time :")
print(df_cleaned['Departure_Time'].head())
```

```python
# Task 2.4 - Save Validated Dataset

df_cleaned.to_csv(
    "/home/jovyan/work/cleaned_data.csv",
    index = False 
)
print("Validated Dataset saved Successfully")
```
